```python
import numpy as np
import lxml.etree as ET
import h5py
import matplotlib.pyplot as plt
```

```python
def read_xdmf(xdmf_file):
    tree = ET.parse(xdmf_file)
    root = tree.getroot()
    # Find the HDF5 file path and dataset path
    h5_file_path = root.find(".//DataItem[@Format='HDF']").text.split(':')[0].strip()
    h5_dataset_path = root.find(".//DataItem[@Format='HDF']").text.split(':')[1].strip()

    # Read the HDF5 file and dataset
    with h5py.File(h5_file_path, 'r') as h5_file:
        data = np.array(h5_file[h5_dataset_path])

    return data
```

```python

xdmf_file = './Datatry.xmf'
data = read_xdmf(xdmf_file)
```
